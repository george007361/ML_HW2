```python
import numpy as np
import pandas as pd
import collections
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
import re
import string
from pymorphy2 import MorphAnalyzer
import nltk
from nltk.corpus import stopwords
from string import punctuation

train_df = pd.read_csv("train.csv")
test_df = pd.read_csv("test.csv")
train_df_cleared = train_df
```

```python
# Получение текстовой строки из списка слов
def str_corpus(corpus):
    str_corpus = ''
    for i in corpus:
        str_corpus += ' ' + i
    str_corpus = str_corpus.strip()
    return str_corpus

# Получение списка всех слов в корпусе
def get_corpus(data):
    corpus = []
    for phrase in data:
        for word in phrase.split():
            corpus.append(word)
    return corpus

# Получение облака слов
def get_wordCloud(corpus):
    wordCloud = WordCloud(background_color='white',
                              stopwords=STOPWORDS,
                              width=3000,
                              height=2500,
                              max_words=200,
                              random_state=42
                         ).generate(str_corpus(corpus))
    return wordCloud

def remove_punct(text):
    for punctuation in string.punctuation:
        text = text.replace(punctuation, '')
    return text

def lemm(doc):
    doc = re.sub(patterns, ' ', doc)
    tokens = []
    for token in doc.split():
        if token:
            token = token.strip()
            token = morph.normal_forms(token)[0]
            
            tokens.append(token)
    if len(tokens) > 2:
        return tokens
    return None 

def preproc(df):
    nltk.download("stopwords")
    russian_stopwords = stopwords.words("russian")
    eng_stopwords = stopwords.words("english")


    # Удаление знаков пунктуации из текста
    df['title'].apply(remove_punct)

    # lower case 
    df['title'] = df['title'].map(lambda x: x.lower())

    # del some words
    banned = {'страница', 'поиск', "купить", "онлайн", "фото", "видео", "тыс", "результатов", "скачать", "смотреть", "mp3", "mail" ,"ru", "млн"}
    df['title'] = df['title'].map(lambda x: x.split(' '))
    df['title'] = df['title'].map(lambda x: [token for token in x if token not in russian_stopwords and token != " " and token.strip() not in punctuation and token not in eng_stopwords and token not in banned])
    df['title'] = df['title'].map(lambda x: ' '.join(x))

    # Лемматизация 
    patterns = "[A-Za-z0-9!#$%&'()*+,./:;<=>?@[\]^_`{|}~—\"\-]+"
    morph = MorphAnalyzer()

    df['title'].apply(lemm)
    return df

def draw(df, count):
    corpus = get_corpus(df['title'].sample(count).values)
    procWordCloud = get_wordCloud(corpus)
    fig = plt.figure(figsize=(20, 8))
    plt.subplot(1, 2, 1)
    plt.imshow(procWordCloud)
    plt.axis('off')
    plt.subplot(1, 2, 1)
```

```python
train_df_cleared = preproc(train_df_cleared)
```

```python
draw(train_df_cleared, 13000)
```

```python
from sklearn.model_selection import train_test_split
X = train_df_cleared['title'].values
y = train_df_cleared['target'].astype(int).values
X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                    train_size=0.8, 
                                                    random_state=42)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import f1_score
from sklearn.feature_extraction.text import TfidfTransformer

tfidf_transformer = TfidfTransformer()
vectorizer = CountVectorizer(max_features = 30000, min_df=4, max_df=0.5)
model = MultinomialNB()

X_train_vec = vectorizer.fit_transform(X_train)
X_train_tfidf = tfidf_transformer.fit_transform(X_train_vec)
print(X_train_tfidf.shape)
model.fit(
    X_train_tfidf,
    y_train
)
y_pred = model.predict(
    X_train_tfidf
)
print(f1_score(y_train, y_pred))

# X_test_vec = vectorizer.fit_transform(X_test)
# # X_test_tfidf = tfidf_transformer.fit_transform(X_test_vec)
# y_pred = model.predict(
#     X_test_tfidf
# )
# print(f1_score(y_test, y_pred))
```

```python
test_df = pd.read_csv("test.csv")
preproc(test_df)
draw(test_df, 10000)
```
